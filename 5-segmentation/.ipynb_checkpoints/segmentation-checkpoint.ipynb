## Segmentation of toy example

In this exercise, you will have to automatically find how many type of coins are present on the picture `coins.jpg`.
Therefore, read the image as `ubyte` and display it. In order to speed-up the subsequent processes, divide the scale of the image by a factor of 4.

```python
Path = './images/'
name = 'coins.jpg'
Coins_im = imread(Path.__add__(name)) #Reading the image
coins_image_ubyte = img_as_ubyte(Coins_im)  #Image as Ubyte
coins_image_scaled = rescale(coins_image_ubyte,1/4) # Rescaling image by 4
print(coins_image_scaled.shape)
imshow(coins_image_scaled)  #Displaying the image 
plt.title('Coins Re-scaled image ')
```

Subsquently, threshold the image using a global Otsu's thresholding after converting the image into gray scale. Display the image.

```python
#from skimage.morphology import disk
coin_grey=rgb2gray(coins_image_scaled)
#radius = 15
#selem = disk(radius)
#local_otsu = rank.otsu(img, selem)
threshold_global_otsu = threshold_otsu(coin_grey)
print("The threshold is ",threshold_global_otsu)
thresh_im=coin_grey>=threshold_global_otsu
imshow(thresh_im)  #Displaying the image 
plt.title('Threshold image ')
```

Clean the image using some morphological operations, i.e. erosion, dilation, opening, closing ...

```python
radius = 2
selem = disk(radius)
im_open=binary_opening(thresh_im,selem)
plt.figure
imshow(im_open)  #Displaying the image 
plt.title('opened image ')
plt.show()
plt.title('opened image ')
im_close=binary_closing(im_open,disk(5))
imshow(im_close)  #Displaying the image 
plt.title('closed image ')
#opening
plt.show()
```

Using the functions [`label`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.label) and [`regionprops`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.regionprops) determine, the radius of each region and determine the number of type of coins in the image.

1. Show the image with the different labels.
1. How many regions do you find?
1. Display the radius of each region. Draw the corresponding circle on the image.

```python
from matplotlib.patches import Circle
from skimage.draw import circle

# Image Labeling
#im_close=1-im_close
im_labeled, numlabels=label(im_close,return_num=True, background=True)
print("The number of regions in the image are",numlabels-1)
imshow(im_labeled)  #Displaying the image 
plt.title('labeled image ')
plt.show()

# Regions properties
reg_pro=regionprops(im_labeled)
numLabel=0
for i in range (1,numlabels):
    x, y=reg_pro[i].equivalent_diameter,reg_pro[i]
    print (" the diametrr of region {} is {} {}with center {}".format(i, x,y, reg_pro[i].centroid))


imshow(im_close)  #Displaying the image 
plt.title('closed image ')
#opening
plt.show()
```

## Segmentation of markers

Considering the techniques you used in the previous exercise, try to segment different objects in `objets4.jpg` image.

Using segmentation find the number of each object in the image. 

Considering that this is not as straight forward as the previous example, come up with your creative solutions and in case of failar explain the reasons and problems. 

```python

Path = './images/'
name = 'objets4.jpg'
objects_rgb = imread(Path.__add__(name)) #Reading the image
objects = rgb2gray(objects_rgb)

# Display the image
plt.figure()
plt.imshow(objects, cmap = 'Greys_r')
plt.title('Input Image')
```

```python
# Compute a threshold to segment the image and display it
thresh = threshold_otsu(objects)
binary_im = objects <= thresh
plt.figure()
plt.imshow(binary_im, cmap = 'Greys_r')
plt.title('Threshold image')

#combilne the the binary images and edge images
edge_im = feature.canny(objects, sigma = 1)
binary_edges_im = binary_im + edge_im
objects_binary_edges_im = binary_edges_im / np.max(binary_edges_im)
plt.figure()
plt.imshow(objects_binary_edges_im.astype('uint8'), cmap= 'Greys_r')
plt.title('Image with edge and threshold combined')

# To perfectly segment the regions and remove unwanted regions we have to apply morphological operations

Im_morph = objects_binary_edges_im.astype('uint8')

#To avoid the black in side the desired objects apply the dilation and errosion to have zero effect
# finally on the size of the regions NUMBER OF ERROSION=DILATION
for i in range (8):
    Im_morph = binary_dilation(Im_morph)
for i in range (8):
    Im_morph = binary_erosion(Im_morph)
    
#After morphological operation the some regions are get connected to errode the apply erosion
for i in range (2):
    Im_morph = binary_erosion(Im_morph)

#Display segmented image
plt.figure()
plt.imshow(Im_morph, cmap = 'Greys_r')
plt.title('Final Segmented image')

# Compute the labeling of the images to determine the labels in the image

Connected_regions, Num_regions = label(Im_morph, neighbors=None, return_num=True, connectivity=2)

#The number of connected Objects
print('The number of regions in the images', Num_regions)
```
